```python
combinations = list(itertools.product(["Rituximab"], marker_names))
path_data = "C:/Users/janrh/Documents/GitHub/TDA-local-assortativity-paper/Rituximab/Treated"

# calculate pairwise colocalization scores
similarity_measure_df_S3 = create_colocalization_scores(marker_names, path_data, combinations)
similarity_measure_df_S3 = similarity_measure_df_S3.apply(lambda x: np.nan_to_num(x))

similarity_measure_df_S3["Marker_combinations"] = ["CD20" + label[9:] for label in similarity_measure_df_S3.index.to_list()]
similarity_measure_df_S3 = similarity_measure_df_S3.set_index("Marker_combinations")
```

```python
# combine the two colocalization dataframes
df_original = pd.concat([similarity_measure_df_S3.T, similarity_measure_df_S0.T], 
                        keys=['Treated', 'Control']).reset_index().drop(labels=["level_1"], axis=1).set_index("level_0")

if "CD20_CD20" in df_original.index.to_list():
    df_original = df_original.drop(labels=["CD20_CD20"], axis=1)
df = df_original.melt(ignore_index=False)
```

```python
top_pol_mean = df_original.groupby("level_0").get_group("Treated").apply(lambda x: np.mean(x))
least_pol_mean = df_original.groupby("level_0").get_group("Control").apply(lambda x: np.median(x))

differential_2 = top_pol_mean - least_pol_mean
differential_2 = differential_2.apply(lambda x: np.round(x, 5))
```

```python
from scipy.stats import wilcoxon

def statical_test(frist_sample, second_sample):

    statistical_significance = []
    for x_pos, marker_combination in enumerate(df_original.columns):
        scores_marker_1 = df_original.groupby("level_0").get_group(frist_sample)[marker_combination].values[:600]
        scores_marker_2 = df_original.groupby("level_0").get_group(second_sample)[marker_combination].values[:600]

        if max(scores_marker_1) == 0 and min(scores_marker_1) == 0 and max(scores_marker_2) == 0 and min(scores_marker_2) == 0:
            statistical_significance.append(np.nan)
        else:
            test = wilcoxon(x=scores_marker_1, y=scores_marker_2)
            
            p_value = test[1]
            if 0.01 < p_value:
                statistical_significance.append(np.nan)

            elif 0.001 < p_value and p_value < 0.01:
                statistical_significance.append(x_pos)

            elif 0.0001 < p_value and p_value < 0.001:
                statistical_significance.append(x_pos - 0.1)
                statistical_significance.append(x_pos + 0.1)

            else:
                statistical_significance.append(x_pos - 0.2)
                statistical_significance.append(x_pos)
                statistical_significance.append(x_pos + 0.2)

    return statistical_significance
```

```python
sns.set(rc={'figure.figsize':(40, 10)})
sns.set_theme(context="paper", style="whitegrid", font_scale=2.5)

fig = plt.figure(figsize=(35, 12))
#fig.suptitle('Differential colocalization', fontsize=22)

ax = fig.add_subplot(2, 1, 1)
sns.boxplot(data=df, x="Marker_combinations", y="value", hue=df.index.values, showfliers=False, 
            width=0.8, palette=["#BDCCDF", "#E3C29F", "#A2C6C2"], linewidth=1.4)

# adjust axis
plt.xticks(rotation=90)
plt.ylabel("Colocalization scores")

sns.reset_orig()
ax = fig.add_subplot(2, 1, 2)
ax.set_title("Change in colocalization", fontsize=30, x=0.5, y=-2.6)

# get max values of the difference to create color bar values
max_values = max(np.abs(differential_2.values))

# test sigificane of mean shift for Treated and control sample, more dots = more sigificant
statists_S_C = statical_test("Treated", "Control")
sns.scatterplot(x=statists_S_C, y=[0] * len(statists_S_C), color="black", sizes=5)

# plot colorbar indicading the shift of colocalization
im = plt.imshow([differential_2.values], interpolation='none', aspect='equal', 
                cmap=sns.diverging_palette(220, 20, as_cmap=True), vmin=-max_values, vmax=max_values)

# add colorbar
cb = plt.colorbar(im, ax=ax, orientation="horizontal", pad=0.2)
cb.ax.tick_params(labelsize=26)

# remove labels from the axis
ax.set_xticklabels([])
ax.set_yticklabels(["T - C "], fontsize=26)
ax.set_xticks([])
ax.set_yticks([0])

plt.tight_layout()
plt.show()
#plt.savefig("Pairwise_diff_plot_retuximab.jpg")
```
